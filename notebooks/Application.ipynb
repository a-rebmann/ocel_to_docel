```python
import pm4py

log = pm4py.read_xes("../input/BPI_Challenge_2019.xes")
```

```python
cases_to_keep = []
for case_id, events in log.groupby("case:concept:name"):
    if events["Cumulative net worth (EUR)"].nunique()>1:
        cases_to_keep.append(case_id)
print("keeping", len(cases_to_keep), "cases")
log = log[log["case:concept:name"].isin(cases_to_keep)]
pm4py.write_xes(log, "../input/BPI_Challenge_2019_dynamic.xes", case_id_key='case:concept:name')
```
